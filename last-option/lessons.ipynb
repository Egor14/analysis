```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
%config InlineBackend.figure_format = 'svg' 
from pylab import rcParams
rcParams['figure.figsize'] = 9, 6
```

```python
df = pd.read_csv('random_timetable2.csv', sep=';', encoding='cp1251')
df.head()
```

# Информация о форме семестровой отчетности зачет/экзамен

# Количество занятий по дням недели для учебных групп

```python
group = input()
```

```python
sns.countplot(df[df['Номер группы'] == int(group)]['День']);
plt.title('Количество пар для определенной группы по дням недели');
```

# Количество занятий конкретной дисциплины в семестре(неделе)

```python
lesson = input()
```

```python
df[(df['Номер группы'] == int(group)) & (df['Предмет'] == lesson)]['Предмет'].count()
```

# Количество пар в неделю по дисциплинам

```python
df[df['Номер группы'] == int(group)]['Предмет'].value_counts()
```

```python
sns.countplot(df[df['Номер группы'] == int(group)]['Предмет']);
plt.title('Сравнение количества занятий по различным дисциплинам');
```

# Соотношение лекций/практик по дисциплинам учебных групп

```python
group = int(input())
lesson = input()
```

```python
sns.countplot(df[(df['Номер группы'] == group) & (df['Предмет'] == lesson)]['Вид занятия']);
plt.title('Количество пар для определенного преподавателя по дням недели');
```

# Общая занятость факультетов

```python
sns.countplot(df['День']);
plt.ylabel('Количество пар');
plt.title('Загруженность факультета');
```

# Общее соотношение лекций/практик на факультете

```python
sns.countplot(x = 'Предмет', hue = 'Вид занятия', data = df);
plt.ylabel('Количество пар');
plt.title('Соотношение лекций и практик по предметам');
```
