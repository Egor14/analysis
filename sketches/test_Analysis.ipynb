```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
%config InlineBackend.figure_format = 'svg' 
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
```

```python
df = pd.read_csv('random_timetable.csv', sep=';', encoding='cp1251')
df.head(30)
```

```python
df['День'].value_counts()
```

```python
sns.countplot(df['День']);
plt.ylabel('Количество пар');
plt.title('Загруженность факультета');
```

```python
sns.countplot(df['Преподаватель']);
plt.ylabel('Количество пар');
plt.title('Загруженность преподавателей');
```

```python
sns.countplot(x = 'День', hue = 'Вид занятия', data = df);
plt.ylabel('Количество пар');
plt.title('Соотношение лекций и практик по дням недели на факультете');
# Может быть полезно для определения наличия больших лекционных аудиторий и небольших кабинетов в конкретный день недели
```
